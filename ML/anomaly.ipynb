```python
from sklearn.ensemble import IsolationForest
import numpy as np
import matplotlib.pyplot as plt

import pandas as pd
```

```python

data = pd.DataFrame({
    "amount": [500, 1100, 1200, 45000, 650, 800000],
    "sa":[10,15,1500,5,3,9]

})
```

**n_estimators** = cont of tree


**contamination** =  % of anomalies

```python

model=IsolationForest(n_estimators=100, max_samples='auto', contamination=float(0.5), max_features=2)


model.fit(data)

prediction=model.predict(data)
print(prediction)

```

```python
rng = np.random.RandomState(42)

# Normal data
normal_data = 0.3 * rng.randn(100, 3)

# Anomalies (far away)
anomalies = rng.uniform(low=4, high=6, size=(5, 3))

# Combine
data = np.vstack([normal_data, anomalies])

df = pd.DataFrame(data, columns=["Feature1", "Feature2", "Feature3"])
```

```python
model = IsolationForest(
    n_estimators=200,
    contamination=0.05,
    random_state=42
)

model.fit(df)
```

```python
df["prediction"] = model.predict(df)
```

```python
print(df["prediction"].value_counts())

# -1 = anomalies
# 1= normal
```

```python
df["score"] = model.decision_function(df[["Feature1","Feature2","Feature3"]])
df['score']
```

```python
df.sort_values("score").head(10)
```

```python
plt.figure(figsize=(8,6))

normal = df[df["prediction"] == 1]
anomaly = df[df["prediction"] == -1]

plt.scatter(normal["Feature1"], normal["Feature2"], label="Normal")
plt.scatter(anomaly["Feature1"], anomaly["Feature2"], color="red", label="Anomaly")

plt.legend()
plt.title("Isolation Forest Detection")
plt.show()
```
